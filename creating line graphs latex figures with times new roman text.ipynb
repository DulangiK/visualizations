```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
plt.rc('xtick', labelsize=14) 
plt.rc('ytick', labelsize=14) 
```

```python
# accuracy
x = [1,2,3,4,5,6,7]
# naive bayes accuracy with semester gpa and all other attributes
y1 = [0.386,0.386,0.41,0.404,0.399,0.41,0.45]
# naive bayes accuracy with z-score and semester gpa
y2=[0.561,0.694,0.799,0.844,0.873,0.873,0.913]
# svm accuracy with semester gpa and all other attributes
svm1=[0.5497,0.5497,0.5732,0.585,0.585,0.5791,0.5732]
# svm accuracy with z-score and semester gpa
svm2=[0.5095,0.6886,0.8046,0.8614,0.8967,0.8971,0.9023]
plt.rcParams['mathtext.fontset'] = 'cm'
plt.rcParams['font.family'] = 'STIXGeneral'

plt.plot(x, y1, label = "NB, m1", color='black')

plt.plot(x, y2, label = "NB,m2", color='gray')
plt.plot(x, svm1, label = "SVM, m1", color='black',linestyle = '-.')
plt.plot(x, svm2, label = "SVM, m2", color='black', linestyle = '--')

#plt.legend(fontsize=14,bbox_to_anchor=(0.27, 0.77))


#plt.legend(fontsize=14,bbox_to_anchor=(0.35, 0.77))

plt.legend(loc='upper center', bbox_to_anchor=(0.5, 1.15),
          fancybox=True, shadow=True, ncol=5)
#plt.title('Accuracy of the models based on different attributes')
plt.xlabel('semester',fontsize=14)
plt.ylabel('accuracy',fontsize=14)

#plt.show()
plt.savefig('/home/dulangi/Documents/prediction analysis/filename2.png', dpi=1200)

```
