```python
# Import libraries, batch 2020/2021
from matplotlib import pyplot as plt
import numpy as np
  
  
# Creating dataset
subjects = ['Arts','Bio-science','Commerce','Physical science','Technology']
  
data = [6.7, 26.7, 20, 26.7, 20]


# Create a set of colors
colors = ['khaki', 'tan', 'palevioletred' ,'thistle', '#8EB897']

# Use it thanks to the color argument
fig = plt.figure(figsize =(10, 7))
plt.pie(data, labels=subjects, labeldistance=1.15, wedgeprops = { 'linewidth' : 1, 'edgecolor' : 'white' }, colors=colors, autopct='%1.1f%%', textprops={'fontsize': 20,'family':'STIXGeneral'});
    
    
    
    

  
# show plot
plt.show()
```
